```python
import os
from PIL import Image
list_of_op=['drought','non_drought']
```

```python
for i in list_of_op:
    
    path = "train_path/%s/"%i
    print(path)
    #os.makedirs(path + 'resized/')
    
    dirs = os.listdir( path )
    def resize():
        c=0
        for item in dirs:
            print(item)
            if os.path.isfile(path+item):
                
                try:
                    print(path+item)
                    im = Image.open(path+item).convert('RGB')
                except:
                    print("except")
                    continue
                f, e = os.path.splitext(path+item)
                print(f,e)
                imResize = im.resize((128,128), Image.ANTIALIAS)
                
                imResize.save(path + 'resized/' + str(c)+".jpg" , 'JPEG', quality=90)
                c=c+1
    resize()
```
